```python
from scipy.stats import ttest_rel
#data=data.dropna()
male=data[data["gender"]=="M"]["hsc_p"]
male1=data[data["gender"]=="M"]["ssc_p"]
ttest_rel(male,male1)
```

```python
#statement:Test whether ssc_p male and hsc_p male are same
#Null Hyposthesis: There is no similarity 10th pass mark of male and 12th pass mark of male
#Reject null hyposthesis if the pvalue is less that 0.05%
# Accept null- hyposthesis and reject- Alternate hypothesis
```

# ANOVA - Analysis of Variance -One way classification

```python
import scipy.stats as stats
stats.f_oneway(data["ssc_p"],data["hsc_p"],data["degree_p"])
# Hypothesis testing- As the pvalue is greater than 0.05 we can accept alternate hypothesis and reject null hypothesis
```

# Two way classification

```python
import statsmodels.api as sm
from statsmodels.formula.api import ols
two_way=ols('sl_no~ssc_p+hsc_p+degree_p+ssc_p:hsc_p:degree_p',data=data).fit()
two=sm.stats.anova_lm(two_way,type=2)
two
```

```python
#Result of Two way classification
#statement:There is significant differnce between ssc_p,hsc_p,degree_P
#Interaction =ssc_p*hsc_p*degree_p
#Ho=There is no significant differnce betwween ssc_p,hsc_p,degree_P
#H1=There is significant differnce betwween ssc_p,hsc_p,degree_P
```
